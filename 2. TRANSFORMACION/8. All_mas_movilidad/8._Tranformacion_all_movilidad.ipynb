```
estacionesMunicipioDF = estacionesMunicipioDF.withColumn("codigo_municipio_m",F.expr("substring(codigo_municipio, 1, length(codigo_municipio)-1)"))
```

```
df.count()
```

```
#Unimos la movilidad con las estaciones mas cercanas
movilidadDF = movilidadDF.alias("df1").join(estacionesMunicipioDF.alias("df2"), F.col("df1.cod_mun_ine")==F.col("df2.codigo_municipio_m"),"inner").select(["df1.fecha","df2.indicativo","df1.total_viajes","df1.total_viajes_km"])
```

```
#Cuantos datos de movilidad hay
movilidadDF.count()
```

```
#
from pyspark.sql import Window
windowSpecAgg  = Window.partitionBy("fecha", "indicativo")

movilidadDF_agg = movilidadDF.withColumn("total_viajes_estacion", F.sum(F.col("total_viajes")).over(windowSpecAgg)) \
.withColumn("total_viajes_km_estacion", F.sum(F.col("total_viajes_km")).over(windowSpecAgg))
movilidadDF_agg_unique = movilidadDF_agg.dropDuplicates(["fecha", "indicativo"])
movilidadDF_agg_unique.persist().count()
```

```
df_all = df.alias("df1").join(movilidadDF_agg_unique.alias("df2"), (F.col("df1.indicativo")== F.col("df2.indicativo")) & (F.col("df1.fecha")== F.col("df2.fecha")), "left").select(["df1.*","df2.total_viajes_estacion","df2.total_viajes_km_estacion"])
df_all.persist().count()
```
