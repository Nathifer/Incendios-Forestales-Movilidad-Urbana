```
alt_pend_risc_estacionDF = alt_pend_risc.alias("df1").join(cod_estacion_municDF.alias("df2"), (F.col("df1.cod_idescat")== F.col("df2.codigo_municipio")), "inner").select(["df1.*", "df2.indicativo"])
display(alt_pend_risc_estacionDF)
```

```
#borrar
vinyolsAltPendDF=alt_pend_risc_estacionDF.filter(F.col("indicativo")=="U6")
vc_ocurrenciaVinyols=vc_ocurrencia.filter((F.col("municipio")=="Vinyols i els Arcs")& (F.col("fecha")=="2016-07-30"))


display(vinyolsAltPendDF.alias("df1").join(vc_ocurrenciaVinyols.alias("df2"), (F.col("df1.indicativo")== F.col("df2.indicativo")), "inner").select(["df2.*","df1.altitud", "df1.porcentaje_pendiente_total", "ALT_RISC", "PERILL", "VULNER","df1.municipio"]))
```

```
display(alt_pend_risc_estacionDF.filter(F.col("indicativo")=="U6"))
display(vc_ocurrencia.filter((F.col("municipio")=="Vinyols i els Arcs")& (F.col("fecha")=="2016-07-30")))
display(vc_ocurrencia.filter((F.col("indicativo")=="U6")& (F.col("fecha")=="2016-07-30")))
```

```
alt_pend_risc_estacionDF.count()
```

```
Rif_cat = alt_pend_risc_estacionDF.alias("df1").join(vc_ocurrencia.alias("df2"), (F.col("df1.indicativo")== F.col("df2.indicativo")), "inner") \
.select(['df2.fecha',
 'df2.indicativo',
 'df2.nombre',
 'df2.provincia',
 'df2.prec',
 'df2.tmax',
 'df2.tmed',
 'df2.tmin',
 'df2.velmedia',
 'df2.racha',
 'df2.sol',
 'df2.presMax',
 'df2.presMin',
 'df2.rhum',
 'df2.ocurrencia', "df1.municipio","df1.cod_idescat","df1.comarca","df1.altitud", "df1.porcentaje_pendiente_total", "ALT_RISC", "PERILL", "VULNER"])
```

```
Rif_cat.count()
```

```
display(Rif_cat.filter((F.col("indicativo")=="U6") & (F.col("fecha")=="2016-07-30")))
```

```
display(Rif_cat.groupBy("fecha","municipio").count().filter("count > 1"))
```

```
Rif_cat = Rif_cat.withColumnRenamed("nombre","nombre_estacion") \
.withColumnRenamed("NOM_MUNI","nombre_municipio") \
.withColumnRenamed("COD_INE","codi_ine") \
.withColumnRenamed("ALT_RISC","alto_riesgo") \
.withColumnRenamed("VULNER","vulner") \
.withColumnRenamed("PERILL","peligro") \
.withColumnRenamed("NOM_COMAR","nombre_comarca") \
.withColumnRenamed("MUNICIPI","municipio_riesgo") \
.withColumnRenamed("cod_idescat","cod_municipio")
```

```
display(Rif_cat)
```

```
Rif_cat = Rif_cat.select(
  ['df2.fecha',
 'df2.indicativo',
      "df1.municipio",
   "cod_municipio",
 'df2.prec',
 'df2.tmax',
 'df2.tmed',
 'df2.tmin',
 'df2.velmedia',
 'df2.racha',
 'df2.sol',
 'df2.presMax',
 'df2.presMin',
 'df2.rhum',
   "df1.altitud", 
   "df1.porcentaje_pendiente_total", "alto_riesgo", "peligro", "vulner","ocurrencia"])
```

```
display(Rif_cat)
```

```
Rif_cat.write.mode("overwrite").parquet(f"/mnt/IncendiosForestalesCAT/prep/riesgo_incendios_catalunya_all")
display(Rif_cat)
```
